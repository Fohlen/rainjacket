```python
import numpy as np

forecasts_sun = np.load("../data/forecasts_sun.npy")
forecasts_rain = np.load("../data/forecasts_rain.npy")
measures_sun = np.load("../data/measures_sun.npy")
measures_rain = np.load("../data/measures_rain.npy")
```

```python
import matplotlib.pyplot as plt


plt.plot(forecasts_sun[209:233], c="b", label="Forecast")
plt.plot(measures_sun[209:233], c="r", label="Measured")
plt.xlabel("Hours since start of recording")
plt.ylabel("Temperature K")
plt.legend()
plt.savefig("forecast_vs_real_sun_daily.png")
plt.show()
```

```python
from scipy.stats import linregress

print(linregress(forecasts_sun[209:233], measures_sun[209:233]))
```
